```python
# map Method
def my_square_method(num):
    return num**2
print(list(map(my_square_method, [3, 4, 1, 8, 10])))
```

```python
list(map(int, input().split())) # Take Number input as an Array
```

```python
def length_check(string):
    if len(string) % 2 == 0:
        return 'even'
    else:
        return 'odd'
list(map(length_check, ['python', 'nil', 'john']))
```

```python
# filter Method
def check_even(num):
    return num%2 == 0
list(filter(check_even, range(1, 25)))
```

```python
# lambda Expressions
square = lambda num: num**2
print(square(5))

# Uses of Lambda Expresss
# Squre of Numbers
print(list(map(lambda num: num**2, range(0, 20, 2))))

# Printing Even Numbers
print(list(filter(lambda num: num%2==0, range(1, 25))))

# Printing Strings in reverse order
print(list(map(lambda string: string[::-1], ['python', 'nilanjan', 'mango'])))

# Check string length
print(list(map(lambda string: 'even' if len(string)%2==0 else 'odd', ['python', 'nilanjan', 'mango'])))
```
