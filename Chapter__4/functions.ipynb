```python
# Using Function
def my_function(a, b):
    """
    Return 50% of summation of a and b
    """
    print(f"Value of a is {a} & Value of b is {b}")
    return sum((a, b))*0.5
my_function(20, 10)
my_function(b=20, a=10)
```

```python
# Default Arguments (default arguments must follows non-default argument)
def my_function_1(a, b, c=0):
    print(f"Value of a is {a}, Value of b is {b}, Value of c is {c}")
    return a + b +c
my_function_1(1, 9) # Output: 10
my_function_1(9, b=9) # Output: 18
my_function_1(a=9, 9) # Throw SyntaxError: positional argument follows keyword argument
```

```python
# Passing arguments
def my_args_func(*args):
    return ''.join(str(item) for item in args)
print(my_args_func(1, '2', 3, 4, 5, 'nilanjan'))
print(my_args_func(1, 245, 53, '14', 5))
```

```python
# Passing Key-Value pairs
def my_kwargs_func(**kwargs):
    if('fruit' in kwargs):
        print('Fruit found', kwargs['fruit'])
my_kwargs_func(fruit='apple', lang='python')
my_kwargs_func(fruit='mango', lang='java')
my_kwargs_func(lang='cpp', channel='TM')
```

```python
# Using *args and **kwargs
def my_combined_func(*args, **kwargs):
    if(len(args)>0):
        print(f"I love this number {args[0]}")
    if 'lang' in kwargs:
        print(f"my favourite language is {kwargs['lang']}")
my_combined_func(fruit='mango')
my_combined_func(lang='python')
my_combined_func(19, 90, 90, fruit='mango', lang='java')
```
