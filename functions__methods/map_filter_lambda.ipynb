```python
# map Method
```

```python
def square(num):
    return num**2
```

```python
square(3)
```

```python
my_list = [1, 2, 3, 4, 5]
```

```python
for item in map(square, my_list):
    print(item)
```

```python
map(square, my_list) # Generate another List at some Location
```

```python
list(map(square, my_list))
```

```python
list(map(int, input().split())) # Take input as an Array
```

```python
def len_check(string):
    if len(string) % 2 == 0:
        return 'even'
    else:
        return 'odd'
```

```python
list(map(len_check, ['python', 'nil', 'john']))
```

```python
# filter Methods
```

```python
def check_even(num):
    return num%2 == 0
```

```python
my_list = [1, 2, 3, 4, 5, 6, 7, 8]
```

```python
filter(check_even, my_list) # Generate another List at some Location
```

```python
list(filter(check_even, my_list))
```

```python
# lambda Expressions
```

```python
square = lambda num: num**2
```

```python
square(4)
```

```python
# Uses of Lambda Expresss
```

```python
list(map(lambda num: num**2, [1, 2, 3, 4, 5]))
```

```python
list(filter(lambda num: num%2 == 0, [1, 2, 3, 4, 5, 6]))
```

```python
list(map(lambda string: string[::-1], ['python', 'nil', 'john']))
```

```python
list(map(lambda string: 'even' if len(string)%2==0 else 'odd', ['python', 'nil', 'john']))
```
