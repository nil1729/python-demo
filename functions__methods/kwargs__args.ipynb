```python
# *args and **kwargs
```

```python
def my_func(a, b, c=0, d=0):
    """
    Return 5% of summation of a and b
    """
    print(a, b)
    print(sum((a, b)))
    return sum((a, b))*0.05
```

```python
my_func(20, 10)
```

```python
def my_args_func(*args):
    print(args)
    return sum(args)*0.05
```

```python
my_args_func(1, 2, 3, 4, 5)
```

```python
def my_kwargs_func(**kwargs):
    print(kwargs)
    if('fruit' in kwargs):
        print(kwargs['fruit'])
    else:
        pass
```

```python
my_kwargs_func(fruit='apple', lang='python')
```

```python
def my_com_func(*args, **kwargs):
    print(args, kwargs)
    print(f"I love this number {args[0]} and my favourite language is {kwargs['lang']}")
```

```python
my_com_func(1729, 20, 30, fruit='apple', lang='django')
```
