```python
number1 = 30
number2 = input("Please Provide a Number")
try:    
    add(number1, number2)
except:
    print("Add operation can't be done")
print("Some more Codes")
```

```python
number1 = 30
try:
    number2 = int(input("Please Provide a Number: "))
    add(number1, number2)
except:
    print("Add operation can't be done")
else:
    print("Numbers added")
finally:
    print("I always run")
print("Some more Codes... Out side try/except/finally")
```

```python
def add_two(n1, n2):
    print(n1 + n2)
try:
    year = 2020
    name = input("Enter your name: ")
    yob = int(input("Enter your Year of Birth: "))
    add_two(year + name)
except TypeError:
    print("Oops!! There was an Type error")
except ValueError:
    print("Oops!! There was an Value error")
else:
    print("Numbers added")
finally:
    print("I always run")
print("Some more Codes... Out side try/except/finally")
```

```python
def ask_for_int():
    while True:
        try:
            num = int(input("Please provide a Number: "))
        except:
            print("Oops! Looks like you not provide a Number")
            continue
        else:
            print("Your input is ", num)
            break
        finally:
            print("I run always at the end try/except/finally")
            
```

```python
ask_for_int() # Enter a Number
```

```python
ask_for_int()  # Enter a String
```
